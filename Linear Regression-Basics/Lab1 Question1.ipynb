####  Question 2. Initialize an averaging filter as follows:

```python
F= np.ones((3,3), dtype="float") * (1 / 9) #Initializing the avaeraging filter
```

####  Question 3. Pad the image with 1s on all sides. This is done using np.pad() function
####  Question 4. Convolve the filter over the given image- This is done by going over the image Height and in the second loop by going over the width and multiplying image selected part with kernel and adding both.
#### The roi will be centered around the current (x, y)-coordinates of the image . The roi will also have the same size as our kernel , which is critical for the next step.
#### Convolution is performed taking the element-wise multiplication between the roi and kernel , followed by summing the entries in the matrix.

Explanation is from Site https://www.pyimagesearch.com/2016/07/25/convolutions-with-opencv-and-python/

```python
def find_convolution(image,kernel):
    (image_H, image_W) = image.shape[:2]
    (kernel_H, kernel_W) = kernel.shape[:2]
    pad=1
#     pad=(kW-1)//2
#     image = cv2.copyMakeBorder(image, pad, pad, pad, pad,cv2.BORDER_REPLICATE)
    image=np.pad(image, pad,mode='constant',constant_values=1) #Image is padded with ones to that convolved image does not lose dimension
    output = np.zeros((image_H, image_W), dtype="float32")
    for y in np.arange(pad, image_H+pad):
        for x in np.arange(pad, image_W+pad):
            image_selection = image[y - pad:y + pad + 1, x - pad:x + pad +1] #Part of Image is selected
            k = (image_selection*kernel).sum() #Multiplication with averaging Filter and addition of two matrices
            output[y-pad,x-pad]=k         
    return output  
```

```python
convolve_output=find_convolution(img,F)
```

#### Question 5: Plot the original and blurred images side by side.

#### As the Averaging Filter is very small, blur is minimal and not clearly visible

```python
fig, (ax1, ax2) = plt.subplots(1, 2)
fig.suptitle('Image Convolution first time',fontsize=18)
ax1.imshow(img,cmap='gray' ,vmin=0,vmax=255)
ax2.imshow(convolve_output,cmap='gray',vmin=0,vmax=255)
```

#### Question 6. Examine the result of applying the filter multiple times.-After the convolution program is run 10 times it gives a blurry image and the image blurredness keeps on increasing as we repeat the process multiple times


```python
convolve_output=find_convolution(img,F)
convolve_output1=find_convolution(convolve_output,F)
convolve_output2=find_convolution(convolve_output1,F)
convolve_output3=find_convolution(convolve_output2,F)
convolve_output4=find_convolution(convolve_output3,F)
convolve_output5=find_convolution(convolve_output4,F)
convolve_output6=find_convolution(convolve_output5,F)
convolve_output7=find_convolution(convolve_output6,F)
convolve_output8=find_convolution(convolve_output7,F)
convolve_output9=find_convolution(convolve_output8,F)
convolve_output10=find_convolution(convolve_output9,F)
```

####  Plot the original and blurred images side by side.

```python
fig, (ax1, ax2) = plt.subplots(1, 2)
fig.suptitle('Image Convolution after multiple times',fontsize=18)
ax1.imshow(img,cmap='gray' ,vmin=0,vmax=255)
ax2.imshow(convolve_output10,cmap='gray',vmin=0,vmax=255)
```
